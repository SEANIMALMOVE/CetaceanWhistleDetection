```python
import pandas as pd

# Ensure Year, Month, Day, Hour, Minute, Second are integers
df['Year'] = df['Year'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Day'] = df['Day'].astype(int)
df['Hour'] = df['Hour'].astype(int)
df['Minute'] = df['Minute'].astype(int)
df['Second'] = df['Second'].astype(int)

# create DateTime column
df["DateTime"] = df.apply(
    lambda row: pd.Timestamp(
        year=row['Year'].astype(int), month=row['Month'].astype(int), day=row['Day'].astype(int),
        hour=row['Hour'].astype(int), minute=row['Minute'].astype(int), second=row['Second'].astype(int)
    ), axis=1
)

# create PathDatetime column
df["PathDatetime"] = df.apply(
    lambda row: (
        pd.Timestamp(
            year=row['Year'], month=row['Month'], day=row['Day'],
            hour=row['Hour'], minute=row['Minute'], second=row['Second']
        ) if row['Second'] >= 24 
        else pd.Timestamp(
            year=row['Year'], month=row['Month'], day=row['Day'],
            hour=row['Hour'], minute=row['Minute'], second=row['Second']
        ) - pd.Timedelta(minutes=1)
    ), axis=1
)

# create Path column
df["Path"] = df.apply(
    lambda row: f"channelA_{row['PathDatetime'].year}-{str(row['PathDatetime'].month).zfill(2)}-{str(row['PathDatetime'].day).zfill(2)}_{str(row['PathDatetime'].hour).zfill(2)}-{str(row['PathDatetime'].minute).zfill(2)}-24.wav",
    axis=1
)
```

## Split Segments

```python
import os
from pydub import AudioSegment
```

```python
# Directorios
input_dir = "../Dataset/Audios/SEANIMALMOVEFull/"  # Carpeta donde están los audios originales
output_dir = "../Dataset/Audios/SEANIMALMOVE/SegmentedAudiosMultiLabel3"  # Carpeta donde guardar los segmentos
os.makedirs(output_dir, exist_ok=True)  # Crear la carpeta si no existe
```

```python
# Frecuencia de muestreo
Fs = 256000  # Hz
```

```python
# Iterar por cada fila del DataFrame
for index, row in final_df.iterrows():
    path = row['Path']

    start_sample = int(Fs * row['StartSecond'])  # Convertir a muestras
    end_sample = int(Fs * (row['EndSecond']))  # Convertir a muestras

    # start_sample y end_sample to miliseconds
    start_ms = start_sample / Fs * 1000
    end_ms = end_sample / Fs * 1000
    
    # Leer el archivo de audio
    audio_path = os.path.join(input_dir, path)
    if not os.path.exists(audio_path):
        print(f"Archivo no encontrado: {audio_path}")
        continue

    audio = AudioSegment.from_file(audio_path)

    # Extraer el segmento
    segment = audio[start_ms:end_ms]

    # Crear un nombre único para el archivo de segmento
    segment_name = f"{os.path.splitext(path)[0]}_{row['StartSecond']}_{row['EndSecond']}.wav"
    output_path = os.path.join(output_dir, segment_name)

    # Exportar el segmento
    segment.export(output_path, format="wav")
    # print(f"Segmento guardado: {output_path}")
```

```python
# Check if all files are created and they duration is 3 seconds
segmented_files = os.listdir(output_dir)
segmented_files = [file for file in segmented_files if file.endswith(".wav")]
notfoundfiles = 0
no3secondsfiles = 0

for index, row in final_df.iterrows():
    path = row['Path']
    segment_name = f"{os.path.splitext(path)[0]}_{row['StartSecond']}_{row['EndSecond']}.wav"
    if segment_name not in segmented_files:
        print(f"Segmento no encontrado: {segment_name}")
        notfoundfiles += 1
    # if found open and check duration is 3 seconds
    else:
        audio_path = os.path.join(output_dir, segment_name)
        audio = AudioSegment.from_file(audio_path)
        duration = len(audio) / 1000
        if duration != 3:
            print(f"Segmento {segment_name} no tiene 3 segundos: {duration}")
            no3secondsfiles += 1

print(f"Total de segmentos no encontrados: {notfoundfiles}")
print(f"Total de segmentos no tienen 3 segundos: {no3secondsfiles}")
```

## Split Dataset

```python
import pandas as pd
# Load CSV
df = pd.read_csv("../Dataset/CSVs/DatasetMultiLabel3.csv")
```

```python
# Preparar DataFrame y columnas de clase
class_columns = [col for col in df.columns if col not in ['Path', 'Segment', 'StartSecond', 'EndSecond', 'Class', 'NumClasses', 'SegmentPath']]

# Obtener una tabla auxiliar con información por Path
path_classes = final_df.groupby('Path')[class_columns].sum()
path_classes['NumClasses'] = path_classes[class_columns].apply(lambda row: (row > 0).sum(), axis=1)
```

```python
from sklearn.model_selection import train_test_split

# 1. Crear un DataFrame que agrupe por `Path` y las clases presentes en cada archivo
path_classes = df.groupby('Path')[class_columns].sum()

# 2. Asignar clases poco representadas completamente al conjunto de entrenamiento
for cls in class_columns:
    paths_with_class = path_classes[path_classes[cls] > 0]
    if len(paths_with_class) < 3:  # Si una clase tiene menos de 2 Paths
        print(f"La clase {cls} no tiene suficientes Path para estratificación (Paths disponibles: {len(paths_with_class)}). Asignando a train.")
        df.loc[df[cls] > 0, 'Subset'] = 'train'

# 3. Filtrar los `Paths` ya asignados a train
preassigned_paths = df[df['Subset'] == 'train']['Path'].unique()
remaining_paths = path_classes[~path_classes.index.isin(preassigned_paths)]

# 4. Dividir los `Paths` restantes entre Train, Validation y Test
#    - Aquí evitamos usar estratificación estricta, pero balanceamos la presencia de clases.
train_paths, test_val_paths = train_test_split(
    remaining_paths.index,
    test_size=0.3,  # 30% para Validación + Prueba
    random_state=42,
)

val_paths, test_paths = train_test_split(
    test_val_paths,
    test_size=0.5,  # 50% de los restantes para Validación y el otro 50% para Prueba
    random_state=42,
)

# 5. Asignar las divisiones a los segmentos originales
df.loc[df['Path'].isin(train_paths), 'Subset'] = 'train'
df.loc[df['Path'].isin(val_paths), 'Subset'] = 'val'
df.loc[df['Path'].isin(test_paths), 'Subset'] = 'test'

# Verificar número de Path que coinciden entre subconjuntos
train_paths_set = set(train_paths)
val_paths_set = set(val_paths)
test_paths_set = set(test_paths)

print(f"Paths en Train y Validación: {len(train_paths_set.intersection(val_paths_set))}")
print(f"Paths en Train y Test: {len(train_paths_set.intersection(test_paths_set))}")
print(f"Paths en Validación y Test: {len(val_paths_set.intersection(test_paths_set))}")
```
