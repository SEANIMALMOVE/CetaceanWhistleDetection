```python
import librosa
import librosa.display

filename = '5400400L.wav'
y, sr = librosa.load(filename)
# trim silent edges
whale_song = y
```

```python
import numpy as np
import matplotlib.pyplot as plt

n_fft = 2048
D = np.abs(librosa.stft(whale_song[:n_fft], n_fft=n_fft, hop_length=n_fft+1))
plt.plot(D);
```

```python
# compute 3 mel spectrograms with diferent hop_length in the same figure

hop_lengths = [64, 512, 4096]
n_mels = 512
fig, ax = plt.subplots(nrows=3, sharex=True, sharey=True, figsize=(12, 8))
for i in range(len(hop_lengths)):
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=256)
    S_DB = librosa.power_to_db(S, ref=np.max)
    img = librosa.display.specshow(S_DB, sr=sr, hop_length=hop_lengths[i], x_axis='time', y_axis='mel', ax=ax[i])
    ax[i].set_title('hop_length = {}'.format(hop_lengths[i]))
fig.colorbar(img, ax=ax, format='%+2.0f dB');
```

```python
# compute 3 mel spectrograms with different n_mels in the same figure
n_mels = [32, 256, 4096]
hop_length = 512
fig, ax = plt.subplots(nrows=3, sharex=True, sharey=True, figsize=(12, 8))
for i in range(len(n_mels)):
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=n_mels[i])
    S_DB = librosa.power_to_db(S, ref=np.max)
    img = librosa.display.specshow(S_DB, sr=sr, hop_length=hop_length, x_axis='time', y_axis='mel', ax=ax[i])
    ax[i].set_title('n_mels = {}'.format(n_mels[i]))
fig.colorbar(img, ax=ax, format='%+2.0f dB');
```
